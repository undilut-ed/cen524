```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
breast_cancer = load_breast_cancer()
```

```python
X = breast_cancer.data
y = breast_cancer.target
```

```python
len(X)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.15)
```

```python
len(Xtrain), len(Xtest)
```

```python
mlpClassifier = MLPClassifier()
```

```python
mlpClassifier.fit(Xtrain, Ytrain)
```

```python
print(f"Accurcay on the train set: {accuracy_score(mlpClassifier.predict(Xtrain), Ytrain) * 100:.02f}%")
```

```python
Yhat = mlpClassifier.predict(Xtest)
```

```python
print(f"Accuracy on the test set: {accuracy_score(Yhat, Ytest) * 100:.02f}%")
```

```python
cm = confusion_matrix(Yhat, Ytest)
```

```python
cm_display = ConfusionMatrixDisplay(cm)
```

```python
cm_display.plot(cmap="Greys");
```
